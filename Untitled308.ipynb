```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#q1
#matplotlib is data visualisation library of pythonn used for data  visualisation to extract more insights from the data 
# it is used for dat visualisation
#bar,histogram,scatter,line,pie
data=pd.read_csv("batter.csv")
```

```python
data.head()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2)

ax[0,0].bar(data['avg'],data['strike_rate'])
ax[0,1].plot(data['avg'],data['strike_rate'])
ax[1,0].plot(data['avg'],data['strike_rate'])
ax[1,1].scatter(data['avg'],data['strike_rate'])
```

```python
#q2
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#q3
#subplots provide us the facility of plotting mutiple data on mutiple graphs together
fig,ax=plt.subplots(nrows=2,ncols=2)
 

ax[0,0].plot( np.array([0, 1, 2, 3, 4, 5]) , np.array([0, 100, 200, 300, 400, 500]))
ax[0,1].plot(np.array([0, 1, 2, 3, 4, 5]),np.array([50, 20, 40, 20, 60, 70]))
ax[1,0].plot(np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60]))
ax[1,1].plot(np.array([0, 1, 2, 3, 4, 5]),np.array([200, 350, 250, 550, 450, 150]))


```

```python
#q4

#bar chart is one of the graph used for data visualisation in matplotlib,bar chart shows the frequency of data.
# it is used for data visualisation
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('company vs profit')

plt.show()



```

```python
#q5
# box plot is used for data visualisation
# it is states how data is distributed among the central tendencies and outliers   
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot(box1)
plt.boxplot(box2)
```
